```python
def write_to_excel(data, file_path, base_sheet_name, number):
    
    # Проверяем, существует ли файл Excel
    try:
        # Пробуем загрузить существующий файл
        book = load_workbook(file_path)
        
    except FileNotFoundError:
        # Если файл не найден, создаем новый
        book = pd.ExcelWriter(file_path, engine='openpyxl') 
        book.book.create_sheet(base_sheet_name)
        book.save()

    # Создаем уникальное имя листа с использованием номера
    sheet_name = f"{base_sheet_name}{number}"

    # Проверяем, существует ли лист с указанным именем
    sheet_exists = sheet_name in book.sheetnames

    # Если лист существует, создаем новый лист с уникальным именем
    if sheet_exists:
        sheet_name = find_unique_sheet_name(book, sheet_name)

    # Записываем данные в Excel
    with pd.ExcelWriter(file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
        if sheet_exists:
            writer.book.create_sheet(sheet_name)
        data.to_excel(writer, sheet_name=sheet_name, index=False)


def find_unique_sheet_name(book, base_name):
    # Находим уникальное имя для нового листа
    suffix = 1
    while f"{base_name}_{suffix}" in book.sheetnames:
        suffix += 1
    return f"{base_name}_{suffix}"

# данные для записи
file_path = r'C:\Users\volod\OneDrive\Рабочий стол\BEEZON LOGISTICS\Февраль.xlsx'
base_sheet_name = 'Февраль_'

# название определяется по дню отправки
number = data['Дата'].dt.day[0]

# запишем в файл
write_to_excel(data, file_path, base_sheet_name, number)
```
